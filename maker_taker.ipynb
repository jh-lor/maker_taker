```python
from maker_taker_base import Game, Taker, Maker, Side
from sample_traders import *
import pandas as pd
import statistics
```

```python
from dataclasses import dataclass


@dataclass
class SimResults:
    mean_pnl: float
    std_pnl: float
    tstat: float
    maker_pnl: list

def run_one_sim(taker, maker, seed = 0):
    return Game(taker, maker).run_game()

def run_sims(taker, maker, n = 100):
    maker_pnl = []
    for sim in range(n):
        maker_pnl.append(run_one_sim(taker,maker)['maker_pnl'])
    return SimResults(statistics.mean(maker_pnl), statistics.stdev(maker_pnl), statistics.mean(maker_pnl)/statistics.stdev(maker_pnl), maker_pnl)
    
```

```python
run_sims(GreedyTaker, CenterMaker, n = 10000)
```
